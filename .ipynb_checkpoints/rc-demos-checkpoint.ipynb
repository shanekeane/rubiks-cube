```python
import numpy as np
import matplotlib.pyplot as plt
import random
from copy import deepcopy
import pandas as pd
```

```python
#Define cube turns
def turn_face_clockwise(face):
    i = [[2,1,0],[2,1,0],[2,1,0]]
    j = [[0,0,0],[1,1,1],[2,2,2]]
    return face[i,j]

def turn_face_anticlockwise(face):
    i = [[0,1,2],[0,1,2],[0,1,2]]
    j = [[2,2,2],[1,1,1],[0,0,0]]
    return face[i,j]   

#All possible moves. These are consistent with Western Colour Scheme.
def F(cube_orig):
    cube = deepcopy(cube_orig)
    cube[0,2,:], cube[1,:,2], cube[5,0,:], cube[3,:,0] = np.flip(cube_orig[1,:,2],0),\
                                                         cube_orig[5,0,:], \
                                                         np.flip(cube_orig[3,:,0], 0), \
                                                         cube_orig[0,2,:]
    cube[2] = turn_face_clockwise(cube_orig[2])
    return cube

def F_prime(cube_orig):
    cube = deepcopy(cube_orig)
    cube[0,2,:], cube[3,:,0], cube[5,0,:], cube[1,:,2] = cube_orig[3,:,0], \
                                                         np.flip(cube_orig[5,0,:], 0), \
                                                         cube_orig[1,:,2], \
                                                         np.flip(cube_orig[0,2,:],0)
    cube[2] = turn_face_anticlockwise(cube_orig[2])
    return cube

def B(cube_orig):
    cube = deepcopy(cube_orig)
    cube[0,0,:], cube[3,:,2], cube[5,2,:], cube[1,:,0] = cube_orig[3,:,2],\
                                                         np.flip(cube_orig[5,2,:], 0), \
                                                         cube_orig[1,:,0], \
                                                         np.flip(cube_orig[0,0,:], 0)
    cube[4] = turn_face_clockwise(cube_orig[4])
    return cube

def B_prime(cube_orig):
    cube = deepcopy(cube_orig)
    cube[0,0,:], cube[3,:,2], cube[5,2,:], cube[1,:,0] = np.flip(cube_orig[1,:,0], 0), \
                                                         cube_orig[0,0,:], \
                                                         np.flip(cube_orig[3,:,2],0), \
                                                         cube_orig[5,2,:]
    cube[4] = turn_face_anticlockwise(cube_orig[4])
    return cube

def L(cube_orig):
    cube = deepcopy(cube_orig)
    cube[0,:,0], cube[2,:,0], cube[5,:,0], cube[4,:,2] = np.flip(cube_orig[4,:,2], 0), \
                                                         cube_orig[0,:,0], \
                                                         cube_orig[2,:,0], \
                                                         np.flip(cube_orig[5,:,0], 0)
    cube[1] = turn_face_clockwise(cube_orig[1])
    return cube

def L_prime(cube_orig):
    cube = deepcopy(cube_orig)
    cube[0,:,0], cube[2,:,0], cube[5,:,0], cube[4,:,2] = cube_orig[2,:,0], \
                                                         cube_orig[5,:,0], \
                                                         np.flip(cube_orig[4,:,2],0), \
                                                         np.flip(cube_orig[0,:,0], 0)
    cube[1] = turn_face_anticlockwise(cube_orig[1])
    return cube

def R(cube_orig):
    cube = deepcopy(cube_orig)
    cube[0,:,2], cube[2,:,2], cube[5,:,2], cube[4,:,0] = cube_orig[2,:,2], \
                                                         cube_orig[5,:,2], \
                                                         np.flip(cube_orig[4,:,0], 0), \
                                                         np.flip(cube_orig[0,:,2], 0)
    cube[3] = turn_face_clockwise(cube_orig[3])
    return cube

def R_prime(cube_orig):
    cube = deepcopy(cube_orig)
    cube[0,:,2], cube[2,:,2], cube[5,:,2], cube[4,:,0] = np.flip(cube_orig[4,:,0], 0),\
                                                         cube_orig[0,:,2], \
                                                         cube_orig[2,:,2], \
                                                         np.flip(cube_orig[5,:,2], 0)
    cube[3] = turn_face_anticlockwise(cube_orig[3])
    return cube

def U(cube_orig):
    cube = deepcopy(cube_orig)
    cube[1,0,:], cube[2,0,:], cube[3,0,:], cube[4,0,:] = cube_orig[2,0,:], \
                                                         cube_orig[3,0,:], \
                                                         cube_orig[4,0,:], \
                                                         cube_orig[1,0,:]
    cube[0] = turn_face_clockwise(cube_orig[0])
    return cube

def U_prime(cube_orig):
    cube = deepcopy(cube_orig)
    cube[1,0,:], cube[2,0,:], cube[3,0,:], cube[4,0,:] = cube_orig[4,0,:], \
                                                         cube_orig[1,0,:], \
                                                         cube_orig[2,0,:], \
                                                         cube_orig[3,0,:]
    cube[0] = turn_face_anticlockwise(cube_orig[0])
    return cube

def D(cube_orig):
    cube = deepcopy(cube_orig)
    cube[1,2,:], cube[2,2,:], cube[3,2,:], cube[4,2,:] = cube_orig[4,2,:], \
                                                         cube_orig[1,2,:], \
                                                         cube_orig[2,2,:], \
                                                         cube_orig[3,2,:]
    cube[5] = turn_face_clockwise(cube_orig[5])
    return cube

def D_prime(cube_orig):
    cube = deepcopy(cube_orig)
    cube[1,2,:], cube[2,2,:], cube[3,2,:], cube[4,2,:] = cube_orig[2,2,:], \
                                                         cube_orig[3,2,:], \
                                                         cube_orig[4,2,:], \
                                                         cube_orig[1,2,:]
    cube[5] = turn_face_anticlockwise(cube_orig[5])
    return cube
```

```python
def generate_solved_cube():
    colours = ('W', 'O', 'G', 'R', 'B', 'Y')
    solved_cube = np.zeros((6, 3, 3), dtype='str')
    for face in range(6):
        solved_cube[face] = colours[face]
    return solved_cube
    
def generate_random_cube():
    possible_moves = (U, U_prime, D, D_prime, L, L_prime, 
                      R, R_prime, F, F_prime, B, B_prime)
    
    cube = generate_solved_cube()
    number_of_moves = np.random.choice(51)+50 #choose 50-100 moves
    moves = random.choices(possible_moves, k=number_of_moves)
    
    for move in moves:
        cube = move(cube)

    return cube
```

```python
def convert_colour_to_numerical(colour_array):
    COLOUR_SERIES = pd.Series(data = [np.array([255,255,255], dtype=np.uint8), 
                                      np.array([183, 18, 52], dtype=np.uint8),
                                      np.array([0, 70, 173], dtype=np.uint8), 
                                      np.array([255,88, 0], dtype=np.uint8),
                                      np.array([0, 155, 72], dtype=np.uint8), 
                                      np.array([255, 213, 0], dtype=np.uint8)],
                          index = ['W', 'R', 'B', 'O', 'G', 'Y'])
    
    out_shape = colour_array.shape + (3,)
    return np.stack(COLOUR_SERIES[colour_array.flatten()].to_numpy(), axis=0, dtype=np.uint8).reshape(out_shape)

def plot_cube(cube):
    fig, ax = plt.subplots(3,4)

    for a in ax.ravel():
        a.tick_params(left=False)
        a.tick_params(bottom=False)
        a.set_xticks([])
        a.set_yticks([])
    
    plt.subplots_adjust(wspace=0, hspace=0)
    ax[0,0].remove()
    ax[0,2].remove()
    ax[0,3].remove()
    ax[2,0].remove()
    ax[2,2].remove()
    ax[2,3].remove()
    ax[0,1].pcolormesh(convert_colour_to_numerical(cube[0]), edgecolors='k', linewidth=0.5)
    ax[0,1].invert_yaxis()
    ax[1,0].pcolormesh(convert_colour_to_numerical(cube[1]), edgecolors='k', linewidth=0.5)
    ax[1,0].invert_yaxis()
    ax[1,1].pcolormesh(convert_colour_to_numerical(cube[2]), edgecolors='k', linewidth=0.5)
    ax[1,1].invert_yaxis()
    ax[1,2].pcolormesh(convert_colour_to_numerical(cube[3]), edgecolors='k', linewidth=0.5)
    ax[1,2].invert_yaxis()
    ax[1,3].pcolormesh(convert_colour_to_numerical(cube[4]), edgecolors='k', linewidth=0.5)
    ax[1,3].invert_yaxis()
    ax[2,1].pcolormesh(convert_colour_to_numerical(cube[5]), edgecolors='k', linewidth=0.5)
    ax[2,1].invert_yaxis()
```

```python
def policy_evaluation(V, pi, rewards, gamma, tol):
    """
    Inputs: pi - policy
            rewards - grid_size x grid_size matrix
            gamma - discount rate
            tol - tolerance (how close successive Vs should be)
    Output: V - state value fn for pi
    """
    delta = 10
    while delta > tol:
        v = deepcopy(V)
        for state in range(0,len(V)-1):
            V[state] = np.sum(pi[state]*(rewards+gamma*v))
        delta = np.sum(np.abs(v-V))
    return V

def policy_improvement(V, rewards, poss_as, gamma, grid_size):
     return get_pi_from_V(V, rewards, poss_as, gamma, grid_size, choose_one=True)
    
def get_actions_from_pi(pi):
    actions = list()
    for i in range(pi.shape[1]-1):
        actions.append(np.random.choice(np.arange(pi.shape[1]), p=pi[i]))
    return np.asarray(actions)
```

```python
#define policy
#defining ps unnecessary as will equal 1 for all relevant cases
#define policy from initial state value function
V = np.zeros((grid_size, grid_size))     
r = np.full((grid_size, grid_size), -1.)
new_best_pi = get_pi_from_V(V, r, gamma, grid_size) #initialize pi
new_pi = get_actions_from_pi(new_best_pi)
```

```python
#define policy and reward
#defining ps unnecessary as will equal 1 for all relevant cases
def get_pi_indices(state):
    #Returns indices for which pi is set to 0.25
    up = state-grid_size
    down = state+grid_size
    left = state-1
    right = state+1
    if up < 0:
        up = state
    if down > grid_size**2-1:
        down = state
    if state%grid_size==grid_size-1:
        right = state
    if state%grid_size==0:
        left = state
    return np.asarray([up, down, left, right])
    
pi = np.zeros((grid_size**2, grid_size**2))
for i in range(1, pi.shape[0]-1):
    non_zero_inds = get_pi_indices(i)
    pi[i][non_zero_inds] = 0.25
    #Deal with corner cases
    if len(np.where(non_zero_inds==i)[0]) > 0:
        pi[i][i] = 0.25*len(np.where(non_zero_inds==i)[0])      
pi[0][0] = 0.5
pi[0][1] = pi[0][grid_size] = 0.25
```

```python
new_best_pi = pi
new_pi = get_actions_from_pi(new_best_pi)
```

```python
#Iterative policy evaluation
old_pi = np.zeros_like(new_pi)

while np.array_equal(new_pi, old_pi) == False:
    old_best_pi = deepcopy(new_best_pi)
    old_pi = get_actions_from_pi(old_best_pi)
    V = policy_evaluation(new_best_pi, rewards, gamma, tol)
    new_best_pi = policy_improvement(V, rewards, gamma, grid_size)
    new_pi = get_actions_from_pi(new_best_pi)
```
